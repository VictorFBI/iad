#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
info = pd.read_csv('Students_info_0.csv')

for i in range(1, 10):
  info_tmp = pd.read_csv('Students_info_' + str(i) + '.csv')
  info = pd.concat([info, info_tmp], axis=0)

marks = pd.read_csv('Students_marks_0.csv')

for i in range(1, 10):
  marks_tmp = pd.read_csv('Students_marks_' + str(i) + '.csv')
  marks = pd.concat([marks, marks_tmp], axis=0)

df = pd.merge(info, marks, on='index', how='left')
df.sample(n=10) # random 10 observations
```

#### Задание 14. Сколько студентов успешно сдали экзамен по математике?

Создайте новый столбец в таблице df под названием Math_PassStatus и запишите в него F, если студент не сдал экзамен по математике (балл за экзамен < passmark), и P иначе.

Посчитайте количество студентов, сдавших и не сдавших экзамен по математике.

Сделайте аналогичные шаги для экзаменов по чтению и письму.

```python
df['Math_PassStatus'] = df['math score'].apply(lambda x: 'F' if x < passmark else 'T')
df['Reading_PassStatus'] = df['reading score'].apply(lambda x: 'F' if x < passmark else 'T')
df['Writing_PassStatus'] = df['writing score'].apply(lambda x: 'F' if x < passmark else 'T')

def calc_statistics(filter: object):
  for name, group in df.groupby(by=filter):
    if name == 'F':
      print("Amount of students who don't pass the exam", end=': ')
    else:
      print("Amount of students who pass the exam", end=' ')
    print(group.shape[0])


calc_statistics('Math_PassStatus')
print()

calc_statistics('Reading_PassStatus')
print()

calc_statistics('Writing_PassStatus')
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
def condition(row):
    if row['Math_PassStatus'] == 'T' and row['Reading_PassStatus'] == 'T' and row['Writing_PassStatus'] == 'T':
        return 'T'
    else:
        return 'F'

df['OverAll_PassStatus'] = df.apply(condition, axis=1)

for name, group in df.groupby(by='OverAll_PassStatus'):
  if name == 'F':
    print("Amount of students who don't pass all exams", end=': ')
  else:
    print("Amount of students who pass all exams", end=' ')
  print(group.shape[0])
  print()
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    [90, 100] = A
####      [80, 90) = B
####      [70, 80) = C
####      [60, 70) = D
####      [50-60) = E
####    [0, 50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(mark: int):
    if (90 <= mark and mark <= 100):
      return 'A'
    if (80 <= mark and mark < 90):
      return 'B'
    if (70 <= mark and mark < 80):
      return 'C'
    if (60 <= mark and mark < 70):
      return 'D'
    if (50 <= mark and mark < 60):
      return 'E'
    return 'F'

def helper(row):
  return GetGrade((row['math score'] + row['reading score'] + row['writing score']) / 3)

df['Grade'] = df.apply(helper, axis=1)
df.sample(n=15)
```
